```python
### Import required packages

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import scipy.signal


#Pull in data
dataset = pd.read_csv('100.csv')

##Will eventually use this file to convert from sample domain to time domain. Not sure the best way to do that right now so holding off on that
##timedomain = pd.read_csv('100annotations.txt')


##Clean up column names (basically just removing spaces. Don't need both MLII AND V5. Doesn't matter which we use, just two different sensors (aka "leads")
dataset.columns = ['SampleNumber', 'MLII', 'V5']

##Going to only take the first 500 rows for now. Easier to work with but this should be able to be scaled up without changing anything
x = dataset[0:500].SampleNumber
y = dataset[0:500].MLII


###Need to smooth data before finding first differences or we'll get way more points than we need due to small variations other than the large peaks we're looking for.
yhat = scipy.signal.savgol_filter(y, 101, 5)


##Create array of "first differences"
firstdiff = []
for i, j in enumerate(yhat[0:-1]):
    firstdiff.append((yhat[i + 1]) - j)




###Find points around where the first difference values change from positive to negative or vice versa
points=[]
for i, j in enumerate(firstdiff[0:-2]):
    if (j>0 and firstdiff[i+1]<0) or (j<0 and firstdiff[i+1]>0):
        points.append(((i - 2), y.values[i - 2]))
        points.append(((i - 1), y.values[i - 1]))
        points.append( (i, y.values[i] ))
        points.append(((i + 1), y.values[i + 1]))
        points.append(((i + 2), y.values[i + 2]))
    else:
        pass


###Plot original data
y.plot()
###Plot smoothed data (need to convert it to pandas Series from numpy array to get it to work)
pd.Series(yhat).plot()
###Plot points near critical points
x_val = [x[0] for x in points]
y_val = [x[1] for x in points]
plt.plot(x_val,y_val,'or')

###Show plot
plt.show()



###Use polynomial regression to fit line to those points


###Traverse regression line and find points where first derivative is 0 or undefined. These should be the critical points

```
